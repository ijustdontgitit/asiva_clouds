```python
#add image transformations to expand dataset
from PIL import Image
import numpy as np
import glob
from random import randrange
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
size = 8 #side length of kernel from center. 16x16 for size=8
maxSqrSize = 145
#midx = 280
#midy = 252
midx = 510/2
midy = 506/2
```

```python
count = 0

for filename in glob.glob('./test2/*.png'):
    if(count ==5):
        break
    im = Image.open(filename)
    oldfilename = filename.split('.')[1].split('/')[2]
    for i in range(1,6):
        #determine a random location to probe a 16x16 kernel from centralized
        x = randrange(midx-maxSqrSize + size, midx+maxSqrSize - size)
        y = randrange(midy-maxSqrSize + size, midy+maxSqrSize - size)
        imk = im.crop((x-size, y-size, x+size, y+size))

        newfilename = './test2/k/' + oldfilename + "_" + str(i) +  '.png'
        imk.save(newfilename)

print('saved images')
```

```python
j = 0
k = 0
bins = 15
i = 0
j = 0

for filename in glob.glob('./test2/k/*.png'):
#    if i ==100:
#        break

    im = Image.open(filename)
    data = np.asarray(im)

    mu = np.average(data)
    sigma = np.std(data)
    
        # normalized distribution 
    histogram, bins = np.histogram(data, bins=15, density=True)
    
    # gaussian
    y = norm.pdf(bins, mu, sigma)
    
    # fundamental stats
    # truncate last element to match up arrays
    y0 = np.delete(y, -1)
    
    absError = histogram - y0
    
    SE = np.square(absError) # squared errors
    MSE = np.mean(SE) # mean squared errors
    RMSE = np.sqrt(MSE) # Root Mean Squared Error, RMSE
    Rsquared = 1.0 - (np.var(absError) / np.var(y0))
    
    #print (absError, SE, MSE, RMSE, Rsquared)

    plot = False
    
#    if ((Rsquared < 0 and sigma < 5) or (Rsquared > 90 and sigma < 15)):
    if (Rsquared < 0 and sigma > 2.5): 
        if plot:
            plt.figure()
    
            plt.subplot(121)
    
            plt.plot(bins, y, 'g', linewidth=2, linestyle='--')
            plt.axvline(mu, color='b', linewidth=2, linestyle='--')
            plt.title(r'$\ \mu=%.3f,\ \sigma=%.3f,\ r2=%.3f$' % (mu, sigma, Rsquared))
            plt.plot(bins[0:-1], histogram)
    
            plt.subplot(122)
    
            plt.title(filename, fontsize=16)
            plt.imshow(data, cmap='gray_r')
            plt.colorbar()
    
            plt.show()
        else:
            x = "cp " + filename + " cloudy/"
            print (x)
        
        j = j + 1
        
    i = i + 1
    
print (i, j)
```
