```python
#add image transformations to expand dataset
from PIL import Image
import glob

for filename in glob.glob('./Images/clear/*.png'):
    im = Image.open(filename)
    im1 = im.transpose(Image.TRANSPOSE)
    im2 = im1.transpose(Image.TRANSPOSE)
    im3 = im2.transpose(Image.TRANSPOSE)
    im4 = im.transpose(Image.FLIP_LEFT_RIGHT)
    im5 = im.transpose(Image.FLIP_TOP_BOTTOM)
    print(filename.split('.')[1] + "rot90" + '.png')
    im1.save('./Images/cloudy/' + filename.split('.')[1].split('\\')[1] + "_rot90" + '.png')
    im2.save('./Images/cloudy/' + filename.split('.')[1].split('\\')[1] + "_rot180" + '.png')
    im3.save('./Images/cloudy/' + filename.split('.')[1].split('\\')[1] + "_rot270" + '.png')
    im4.save('./Images/cloudy/' + filename.split('.')[1].split('\\')[1] + "_flipLR" + '.png')
    im5.save('./Images/cloudy/' + filename.split('.')[1].split('\\')[1] + "_flipTB" + '.png')
    print('saved images')
```

```python
#organize into test, train, and val directories
from pathlib import Path
from random import shuffle
from glob import glob

files = glob("./Images/cloudy/*.png")
#train, test, val split = (60%, 20%, 20%)
#shuffle(files)
for count, file in enumerate(files):
    if(count <= 0.6 * len(files)):
        path = "./Images/train/cloudy/"
    elif(count <= 0.8 * len(files)):
        path = "./Images/val/cloudy/"
    else:
        path = "./Images/test/cloudy/"
    
    Path(file).rename(path + file.split('.')[1].split('\\')[1] + ".png")
```

```python
#add image transformations to expand dataset
from random import randrange
from PIL import Image
import glob
size = 8 #side length of kernel from center. 16x16 for size=8
maxSqrSize = 145
midx = 280
midy = 252
count = 0
for filename in glob.glob('./Images/val/cloudy/*.png'):
    if(count ==5):
        break
    im = Image.open(filename)
    for i in range(1,6):
        #determine a random location to probe a 16x16 kernel from centralized
        x = randrange(midx-maxSqrSize + size, midx+maxSqrSize - size)
        y = randrange(midy-maxSqrSize + size, midy+maxSqrSize - size)
        imk = im.crop((x-size, y-size, x+size, y+size))
        imk.save('./Kernels_16x16/val/cloudy/' + filename.split('.')[1].split('\\')[1] + "_16x16_" + str(i) +  '.png')
print('saved images')
```
